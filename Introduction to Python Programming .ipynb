```python
# viet 1 cau lenh trong do neu a > b thi in ra cau a greater than b, neu a==b thi in a equal to b, neu a< b in ra cau a smaller than b

a = 10
b = 12

if a > b:
    print('a greater than b')
elif a==b:
    print('a equal to b')
elif a<b:
    print('a smaller than b')
```

```python
# Calculate sum of element in the list

num_list = [4,2,5,2]

sumnum = 0

for element in num_list:
    sumnum = sumnum + element
    
print(sumnum)
```

```python
# vong 1: element = 4, sumnum = 0 -- > sumnum = sumnum + element = 0 + 4 = 4
# vong 2: element = 2, sumnum = 4 -- > sumnum = sumnum + element = 4 + 2 = 6
# vong 3: element = 5, sumnum = 6 -- > sumnum = sumnum + element = 6 + 5 = 11
#vong 4: element = 2, sumnum = 11 -- > sumnum = sumnum + element = 11 + 2 = 13
```

```python
# Sum so chan in the list
num_list = [4,2,5,2]

sumnum = 0

for element in num_list:
    if element%2 == 0:
        sumnum = sumnum + element
    elif element%2 == 1:
        sumnum = sumnum 
print(sumnum)

```

```python
# vong 1: element = 4, sumnum = 0, element chia het 2 -- > sumnum = sumnum + element = 0 + 4 = 4
# vong 2: element = 2, sumnum = 4, element chia het 2 -- > sumnum = sumnum + element = 4 + 2 = 6
# vong 3: element = 5, sumnum = 6, element ko chia het 2 -- > sumnum = sumnum  = 6
# vong 4: element = 2, sumnum = 6, element chia het 2 -- > sumnum = sumnum + element = 6+2 =8
```
